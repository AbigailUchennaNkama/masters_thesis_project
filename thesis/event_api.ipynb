```python
import json
import pandas as pd
from bs4 import BeautifulSoup

def extract_events_from_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    event_links = soup.find_all('a', class_='event-card-link')
    
    events = []
    for link in event_links:
        # Extract the URL
        url = link['href'] if link.get('href') else 'N/A'
        
        # Extract the event name from aria-label
        event_name = link.get('aria-label', 'N/A').replace("View ", "").strip()
        
        # Extract other attributes
        event_location = link.get('data-event-location', 'N/A')
        event_id = link.get('data-event-id', 'N/A')
        event_category = link.get('data-event-category', 'N/A')
        event_paid_status = link.get('data-event-paid-status', 'N/A')
        
        events.append({
            "Event Name": event_name,
            "Event URL": url,
            "Event Location": event_location,
            "Event ID": event_id,
            "Event Category": event_category,
            "Event Paid Status": event_paid_status
        })
    
    return events

# Load the HAR file
with open('/home/nkama/masters_thesis_project/thesis/eventbrite-data_2.har') as f:
    har_data = json.load(f)

all_events = []

# Process all relevant entries in the HAR file
for entry in har_data['log']['entries']:
    if 'text/html' in entry['response']['content'].get('mimeType', ''):
        try:
            html_content = entry['response']['content']['text']
            events = extract_events_from_html(html_content)
            all_events.extend(events)
        except Exception as e:
            print(f"Error processing entry: {e}")

# Create a DataFrame from all extracted events
df = pd.DataFrame(all_events)

# Remove duplicates based on Event ID
df = df.drop_duplicates(subset='Event ID')

# Print the DataFrame
print(df)
print(f"Total unique events extracted: {len(df)}")

# Save to CSV (optional)
#df.to_csv('eventbrite_events.csv', index=False)
#print(f"Saved {len(df)} events to eventbrite_events.csv")
```

```python
import requests
import pandas as pd
import time
import os
from dotenv import load_dotenv

load_dotenv()

# Configuration
EVENTBRITE_TOKEN = os.getenv('EVENTBRITE_TOKEN')  # Set in environment variables
MAX_RETRIES = 3
REQUEST_DELAY = 1  # Seconds between requests to avoid rate limiting

def fetch_event_details(event_id):
    """Fetch detailed event information from Eventbrite API"""
    endpoint = f"https://www.eventbriteapi.com/v3/events/{event_id}/"
    headers = {"Authorization": f"Bearer {EVENTBRITE_TOKEN}"}

    for attempt in range(MAX_RETRIES):
        try:
            response = requests.get(endpoint, headers=headers, timeout=10)
            response.raise_for_status()
            return response.json()

        except requests.exceptions.HTTPError as err:
            if response.status_code == 401:
                print(f"Authentication failed for event {event_id}: {err}")
                return None
            if response.status_code == 404:
                print(f"Event {event_id} not found")
                return None
            else:
                 print(f"HTTP error {response.status_code} for event {event_id}: {err}")
                 return None
        except (requests.exceptions.RequestException, ValueError) as err:
            print(f"Error fetching event {event_id}: {err}")
            time.sleep(2 ** attempt)  # Exponential backoff

    return None
# #fetch_event_details("130323899291")
def clean_dataframe(df):
    """
    Cleans the DataFrame by filling missing values with empty strings,
    handling NaT datetime values, and stripping whitespace from string columns.
    """
    # Fill NaN values with empty string
    df = df.fillna('')

    # Convert datetime columns
    date_cols = df.select_dtypes(include=['datetime64[ns, UTC]']).columns
    for col in date_cols:
        df[col] = df[col].astype(str).replace('NaT', '')

    # Identify object (string) columns
    str_cols = df.select_dtypes(include=['object']).columns
    # Strip whitespace from object columns
    for col in str_cols:
        df[col] = df[col].str.strip()

    return df

#df is supposed to come from:

import json
import pandas as pd
from bs4 import BeautifulSoup

def extract_events_from_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    event_links = soup.find_all('a', class_='event-card-link')
    
    events = []
    for link in event_links:
        # Extract the URL
        url = link['href'] if link.get('href') else 'N/A'
        
        # Extract the event name from aria-label
        event_name = link.get('aria-label', 'N/A').replace("View ", "").strip()
        
        # Extract other attributes
        event_location = link.get('data-event-location', 'N/A')
        event_id = link.get('data-event-id', 'N/A')
        event_category = link.get('data-event-category', 'N/A')
        event_paid_status = link.get('data-event-paid-status', 'N/A')
        
        events.append({
            "Event_Name": event_name,
            "Event_URL": url,
            "Event_Location": event_location,
            "Event_ID": event_id,
            "Event_Category": event_category,
            "Event_Paid_Status": event_paid_status
        })
    
    return events

# Load the HAR file
with open('/home/nkama/masters_thesis_project/thesis/eventbrite_events.har') as f:
    har_data = json.load(f)

all_events = []

# Process all relevant entries in the HAR file
for entry in har_data['log']['entries']:
    if 'text/html' in entry['response']['content'].get('mimeType', ''):
        try:
            html_content = entry['response']['content']['text']
            events = extract_events_from_html(html_content)
            all_events.extend(events)
        except Exception as e:
            print(f"Error processing entry: {e}")

# Create a DataFrame from all extracted events
df = pd.DataFrame(all_events)

# Remove duplicates based on Event ID
df = df.drop_duplicates(subset='Event_ID')

# Print the DataFrame
#print(df)
print(f"Total unique events extracted: {len(df)}")

# Save to CSV (optional)
#df.to_csv('eventbrite_events.csv', index=False)
#print(f"Saved {len(df)} events to eventbrite_events.csv")


event_ids = df["Event_ID"].tolist()
events_data = []

for idx, event_id in enumerate(event_ids[:5]):
    event_details = fetch_event_details(event_id)

    if event_details:
        # Extract comprehensive event data
        event_info = {
            'id': event_details.get('id'),
            'name': event_details.get('name', {}).get('text', ''),
            'description': event_details.get('description', {}).get('text', ''),
            'url': event_details.get('url'),
            'vanity_url': event_details.get('vanity_url'),
            'start_timezone': event_details.get('start', {}).get('timezone'),
            'start_local': event_details.get('start', {}).get('local'),
            'start_utc': event_details.get('start', {}).get('utc'),
            'end_timezone': event_details.get('end', {}).get('timezone'),
            'end_local': event_details.get('end', {}).get('local'),
            'end_utc': event_details.get('end', {}).get('utc'),
            'organization_id': event_details.get('organization_id'),
            'created': event_details.get('created'),
            'changed': event_details.get('changed'),
            'published': event_details.get('published'),
            'capacity': event_details.get('capacity'),
            'capacity_is_custom': event_details.get('capacity_is_custom'),
            'status': event_details.get('status'),
            'currency': event_details.get('currency'),
            'listed': event_details.get('listed'),
            'shareable': event_details.get('shareable'),
            'online_event': event_details.get('online_event'),
            'tx_time_limit': event_details.get('tx_time_limit'),
            'hide_start_date': event_details.get('hide_start_date'),
            'hide_end_date': event_details.get('hide_end_date'),
            'locale': event_details.get('locale'),
            'is_locked': event_details.get('is_locked'),
            'privacy_setting': event_details.get('privacy_setting'),
            'is_series': event_details.get('is_series'),
            'is_series_parent': event_details.get('is_series_parent'),
            'inventory_type': event_details.get('inventory_type'),
            'is_reserved_seating': event_details.get('is_reserved_seating'),
            'show_pick_a_seat': event_details.get('show_pick_a_seat'),
            'show_seatmap_thumbnail': event_details.get('show_seatmap_thumbnail'),
            'show_colors_in_seatmap_thumbnail': event_details.get('show_colors_in_seatmap_thumbnail'),
            'source': event_details.get('source'),
            'is_free': event_details.get('is_free'),
            'version': event_details.get('version'),
            'summary': event_details.get('summary'),
            'facebook_event_id': event_details.get('facebook_event_id'),
            'logo_id': event_details.get('logo_id'),
            'organizer_id': event_details.get('organizer_id'),
            'venue_id': event_details.get('venue_id'),
            'category_id': event_details.get('category_id'),
            'subcategory_id': event_details.get('subcategory_id'),
            'format_id': event_details.get('format_id'),
            'is_externally_ticketed': event_details.get('is_externally_ticketed'),
        }

        def safe_get(d, *keys):
            """Safely get a value from a nested dictionary."""
            for key in keys:
                if isinstance(d, dict):
                    d = d.get(key)
                else:
                    return None
            return d

        # Add logo details if available
        if event_details.get('logo'):
            logo = event_details['logo']
            event_info.update({
                'logo_crop_mask_top_left_x': safe_get(logo, 'crop_mask', 'top_left', 'x'),
                'logo_crop_mask_top_left_y': safe_get(logo, 'crop_mask', 'top_left', 'y'),
                'logo_crop_mask_width': safe_get(logo, 'crop_mask', 'width'),
                'logo_crop_mask_height': safe_get(logo, 'crop_mask', 'height'),
                'logo_original_url': safe_get(logo, 'original', 'url'),
                'logo_original_width': safe_get(logo, 'original', 'width'),
                'logo_original_height': safe_get(logo, 'original', 'height'),
                'logo_id': logo.get('id'),
                'logo_url': logo.get('url'),
                'logo_aspect_ratio': logo.get('aspect_ratio'),
                'logo_edge_color': logo.get('edge_color'),
                'logo_edge_color_set': logo.get('edge_color_set'),
            })




        events_data.append(event_info)

    # Respect rate limits
    if idx < len(event_ids[:5]) - 1:
        time.sleep(REQUEST_DELAY)

# Create DataFrame
events_df = pd.DataFrame(events_data)

# Convert datetime columns
datetime_cols = ['start_utc', 'end_utc', 'created', 'changed']
for col in datetime_cols:
    if col in events_df:
        events_df[col] = pd.to_datetime(events_df[col], errors='coerce')

print(f"Successfully retrieved {len(events_df)} events")
print(events_df.head().to_markdown(index=False))
```
